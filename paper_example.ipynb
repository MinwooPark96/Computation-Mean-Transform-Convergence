$(i)$ Since $T$ is a nilpotent operator of order $2$, we may consider $T=
\begin{pmatrix}
0&\alpha\\
0&0
\end{pmatrix}$ on ${\mathbb C}^2$. 
 Then $T:=U|T|=\begin{pmatrix}
    0 & e^{i\theta} \\
    0  & 0 
    \end{pmatrix}\begin{pmatrix}
    0&0\\
    0&|\alpha|
    \end{pmatrix}$ is the polar decomposition of $T$, where $U=
    \begin{pmatrix}
    0 & e^{i\theta} \\
   0 & 0 
    \end{pmatrix}$ for some  $\theta \in \mathbb{R}$
    and $|T|=
    \begin{pmatrix}
    0&0\\
    0&|\alpha|
    \end{pmatrix}$ for $\alpha=e^{i\theta}|\alpha|\in{\mathbb C}$.
Thus
\begin{align*}
    \widehat{T}=\frac{1}{2}(T+\widetilde{T}^{D})=\frac{1}{2}(U|T|+|T|U)=\frac{1}{2}
    \begin{pmatrix}
    0&\alpha\\
    0&0
    \end{pmatrix}.
\end{align*}

#### Example 2.5.
 If
$T=
\begin{pmatrix}
0&1\\
0&0
\end{pmatrix}$ on ${\mathbb C}^2$, then $T:=U|T|=\begin{pmatrix}
    0 & 1 \\
   0 & 0
    \end{pmatrix}\begin{pmatrix}
    0&0\\
    0& 1
    \end{pmatrix}$ is the polar decomposition of $T$.
Thus
$
    \widehat{T}=\frac{1}{2}(T+\widetilde{T}^{(D)})=\frac{1}{2}
    \begin{pmatrix}
    0&1\\
    0&0
    \end{pmatrix}
$
and so $\widehat{T}$ is normal.
Hence $\widehat{T}^{(n)}$ converges to normal as $n\rightarrow \infty$.

$(ii)$ Suppose that $T$ is idempotent. Then $T$ is unitarily equivalent to $A=
\begin{pmatrix}
0 & R\\
0 & 1
\end{pmatrix}$ for some $R\in\cal{L}(\mathcal{H})$ with $|R^{\ast}|=|R|$. We may consider $\widehat{A}^{(n)}$.
Then $|A|=
\begin{pmatrix}
0 & 0\\
0 & {\sqrt{|R|^2+1}}
\end{pmatrix}$ and $U=
\begin{pmatrix}
0 & \frac{R}{\sqrt{|R|^2+1}}\\
0 & \frac{1}{\sqrt{|R|^2+1}}
\end{pmatrix}
$. Therefore we get that
\begin{align*}
    &U|A|=\begin{pmatrix}
0 & \frac{R}{\sqrt{|R|^2+1}}\\
0 & \frac{1}{\sqrt{|R|^2+1}}
\end{pmatrix}
\begin{pmatrix}
0 & 0\\
0 & {\sqrt{|R|^2+1}}
\end{pmatrix}=
\begin{pmatrix}
0 & R\\
0 & 1
\end{pmatrix}~\mbox{and}\\
&\widetilde{A}^D=|A|U=
\begin{pmatrix}
0 & 0\\
0 & {\sqrt{|R|^2+1}}
\end{pmatrix}
\begin{pmatrix}
0 & \frac{R}{\sqrt{|R|^2+1}}\\
0 & \frac{1}{\sqrt{|R|^2+1}}
\end{pmatrix}=
\begin{pmatrix}
0 & 0\\
0 & 1
\end{pmatrix}.
\end{align*}
Thus 
\begin{align*}
\widehat{A}=\frac{1}{2}(A+\widetilde{A}^D)=\frac{1}{2}
\begin{pmatrix}
0 & R\\
0 &2
\end{pmatrix}=
\begin{pmatrix}
0 & \frac{R}{2}\\
0 & 1
\end{pmatrix}.
\end{align*}

#### Example 2.6.

```python
import time
import random
from tqdm import tqdm

def make_random(option=0,col=2):

    if option == 0: # random number from gaussian distribution
        T_list = [random.gauss(mu=0,sigma=1) for _ in range(col*col)]
        T = np.matrix(T_list).reshape(col,col)
        return T
    elif option == 1: # random number from R
        T_list = [random.random() for _ in range(col*col)]
        T=np.matrix(T_list).reshape(col,col)
        return T

    elif option == 2 : #random number from C
        T_list = [complex(random.random(),random.random()) for _ in range(col*col)]
        T=np.matrix(T_list).reshape(col,col)
        return T

    else:
        raise Exception("not develop")

def compute_random(col=2,n=10,sample_size = 10):

    Transformer = MeanTrans()
    #computing
    for idx in tqdm(range(sample_size)):
        option = random.choice([0,1,2])
        T = make_random(option,col = 2)

        Transformer.transform(T,n)
        Transformer.normal_calculator()

        pre_ch = Transformer.normal_info[0]
        for ch in Transformer.normal_info:
            if pre_ch - ch < 0 and abs(pre_ch - ch) > 1e-14 :
                print("ERROR : EXCEPTION OF NO DECREASE")
                Transformer.normal_plot()
                return
            pre_ch = ch

compute_random(col=2,n=10,sample_size=100000)
```
