#### Theorem 2.3.

Assume that $T$ on ${\mathbb C}^2$ is one of a nilpotent operator of order $2$ and an idempotent operator.
Then $\widehat{T}^{(n)}$ converges to normal as $n\rightarrow \infty$.

#### Example 2.4.
 If
$T=
\begin{pmatrix}
0&1\\
0&0
\end{pmatrix}$ on ${\mathbb C}^2$, then $T:=U|T|=\begin{pmatrix}
    0 & 1 \\
   0 & 0 
    \end{pmatrix}\begin{pmatrix}
    0&0\\
    0& 1
    \end{pmatrix}$ is the polar decomposition of $T$.  
Thus
$
    \widehat{T}=\frac{1}{2}(T+\widetilde{T}^{(D)})=\frac{1}{2}
    \begin{pmatrix}
    0&1\\
    0&0
    \end{pmatrix}
$
and so $\widehat{T}$ is normal. 
Hence $\widehat{T}^{(n)}$ converges to normal as $n\rightarrow \infty$.

$(i)$ Since $T$ is a nilpotent operator of order $2$, we may consider $T=
\begin{pmatrix}
0&\alpha\\
0&0
\end{pmatrix}$ on ${\mathbb C}^2$. 
 Then $T:=U|T|=\begin{pmatrix}
    0 & e^{i\theta} \\
    0  & 0 
    \end{pmatrix}\begin{pmatrix}
    0&0\\
    0&|\alpha|
    \end{pmatrix}$ is the polar decomposition of $T$, where $U=
    \begin{pmatrix}
    0 & e^{i\theta} \\
   0 & 0 
    \end{pmatrix}$ for some  $\theta \in \mathbb{R}$
    and $|T|=
    \begin{pmatrix}
    0&0\\
    0&|\alpha|
    \end{pmatrix}$ for $\alpha=e^{i\theta}|\alpha|\in{\mathbb C}$.
Thus
\begin{align*}
    \widehat{T}=\frac{1}{2}(T+\widetilde{T}^{D})=\frac{1}{2}(U|T|+|T|U)=\frac{1}{2}
    \begin{pmatrix}
    0&\alpha\\
    0&0
    \end{pmatrix}.
\end{align*}

$(ii)$ Suppose that $T$ is idempotent. Then $T$ is unitarily equivalent to $A=
\begin{pmatrix}
0 & R\\
0 & 1
\end{pmatrix}$ for some $R\in\cal{L}(\mathcal{H})$ with $|R^{\ast}|=|R|$. We may consider $\widehat{A}^{(n)}$.
Then $|A|=
\begin{pmatrix}
0 & 0\\
0 & {\sqrt{|R|^2+1}}
\end{pmatrix}$ and $U=
\begin{pmatrix}
0 & \frac{R}{\sqrt{|R|^2+1}}\\
0 & \frac{1}{\sqrt{|R|^2+1}}
\end{pmatrix}
$. Therefore we get that
\begin{align*}
    &U|A|=\begin{pmatrix}
0 & \frac{R}{\sqrt{|R|^2+1}}\\
0 & \frac{1}{\sqrt{|R|^2+1}}
\end{pmatrix}
\begin{pmatrix}
0 & 0\\
0 & {\sqrt{|R|^2+1}}
\end{pmatrix}=
\begin{pmatrix}
0 & R\\
0 & 1
\end{pmatrix}~\mbox{and}\\
&\widetilde{A}^D=|A|U=
\begin{pmatrix}
0 & 0\\
0 & {\sqrt{|R|^2+1}}
\end{pmatrix}
\begin{pmatrix}
0 & \frac{R}{\sqrt{|R|^2+1}}\\
0 & \frac{1}{\sqrt{|R|^2+1}}
\end{pmatrix}=
\begin{pmatrix}
0 & 0\\
0 & 1
\end{pmatrix}.
\end{align*}
Thus 
\begin{align*}
\widehat{A}=\frac{1}{2}(A+\widetilde{A}^D)=\frac{1}{2}
\begin{pmatrix}
0 & R\\
0 &2
\end{pmatrix}=
\begin{pmatrix}
0 & \frac{R}{2}\\
0 & 1
\end{pmatrix}.
\end{align*}

```python
# where R = 2.789 
# R^2 + 1 ~= 8.7785
# 1/sqrt(R^2 + 1) ~= 0.33751
# R/sqrt(R^2 + 1) ~= 0.94132

T=np.matrix([0,2.789,0,1]).reshape(2,2) 
Transformer=MeanTrans()
Transformer.transform(T,10)

print(Transformer[1]) 
Transformer.process(1,dicimals=10)

# sqrt(R^2/4+1) ~= 1.71599
# 1/sqrt(R*R/4+1) ~= 0.58275
# (R/2)/sqrt(R*R/4+1) ~= 0.81264

print(Transformer[2]) 
Transformer.process(2,dicimals=10)
```

#### Example 2.6.

Let $T=
\begin{pmatrix}
0 & 1 \\
0 & 1
\end{pmatrix}$.
Then $T=U|T|=
\begin{pmatrix}
    0 & \frac{\sqrt{2}}{2} \\
   0 & \frac{\sqrt{2}}{2} 
    \end{pmatrix}\begin{pmatrix}
    0&0\\
   0& \sqrt{2}
    \end{pmatrix}$ is the polar decomposition of $T$ and $\widetilde{T}^D=
\begin{pmatrix}
    0&0\\
   0& \sqrt{2}
    \end{pmatrix}
\begin{pmatrix}
    0 & \frac{\sqrt{2}}{2} \\
   0 & \frac{\sqrt{2}}{2} 
    \end{pmatrix}=
\begin{pmatrix}
0 & 0\\
0 & 1\\ 
\end{pmatrix}$. Thus
\begin{align*}
    T_1=\widehat{T}=\frac{1}{2}\left[
    \begin{pmatrix}
0 & 1\\
0 & 1\\ 
\end{pmatrix}+
\begin{pmatrix}
0 & 0\\
0 & 1\\ 
\end{pmatrix}\right]=\frac{1}{2}\begin{pmatrix}
0 & 1\\
0 & 2\\ 
\end{pmatrix}=
\begin{pmatrix}
0 & \frac{1}{2}\\
0 & 1\\ 
\end{pmatrix}.
\end{align*}
And so
\begin{align*}
&|T_1|=(\widehat{T}^*\widehat{T})^\frac{1}{2}
    =\left[
    \begin{pmatrix}
0 & 0\\
\frac{1}{2} & 1\\ 
\end{pmatrix}
\begin{pmatrix}
0 & \frac{1}{2}\\
0 & 1\\ 
\end{pmatrix}\right]^\frac{1}{2}
=
\begin{pmatrix}
0 & 0\\
0 & \frac{\sqrt{5}}{2}\\ 
\end{pmatrix} and
&U_1=\frac{1}{\sqrt{5}}
\begin{pmatrix}
0 & 1\\
0 & 2
\end{pmatrix}.
\end{align*}
Therefore
\begin{align*}
    T_2=\widehat{T_1}&=
    \frac{1}{2}\left[
    \begin{pmatrix}
0 & \frac{1}{2}\\
0 & 1\\ 
\end{pmatrix}+
\begin{pmatrix}
0 & 0\\
0 & \frac{\sqrt{5}}{2}\\ 
\end{pmatrix}
\frac{1}{\sqrt{5}}
\begin{pmatrix}
0 & 1\\
0 & 2\\ 
\end{pmatrix}
\right]\\
&=\frac{1}{2}
\begin{pmatrix}
0 & \frac{1}{2}\\
0 & 2\\ 
\end{pmatrix}=
\begin{pmatrix}
0 & \left(\frac{1}{2}\right)^2\\
0 & 1\\ 
\end{pmatrix}.
\end{align*}

```python
import time
import random
from tqdm import tqdm

def make_random(option=0,col=10):
    current_time = int(time.time())
    random.seed(current_time)

    if option == 0:
        T_list = [random.gauss(mu=0,sigma=1) for _ in range(col*col)]
        T = np.matrix(T_list).reshape(col,col)
        return T

    elif option == 1:
        T_list = [random.random() for _ in range(col*col)]
        T=np.matrix(T_list).reshape(col,col)
        return T
    elif option == 2:
        T_list = [random.random() for _ in range(col*col)]
        T = np.matrix(T_list).reshape(col,col)
        return 1 / (1 +np.exp(-T))

    elif option == 3 :
        T_list = [complex(random.random(),random.random()) for _ in range(col*col)]
        T=np.matrix(T_list).reshape(col,col)
        return T

    elif option == 4:
        T_list = [random.randint(1,100) for _ in range(col*col)]
        T = np.matrix(T_list).reshape(col,col)
        return T
    else:
        raise Exception("not develop")



def compute_bucket(col=10,n=100,sample_size = 10000,iter = 5):

    Transformer = MeanTrans()
    buckets = list()

    #computing
    for idx in range(iter):
        print("=========="+"<start>"+"==========="+"({0}/{1})".format(idx+1,iter))
        bucket=list()

        for _ in tqdm(range(sample_size)):
            T = make_random(option=idx)
            Transformer.transform(T,n)
            Transformer.normal_calculator()

            interest = Transformer.normal_info[0]/Transformer.normal_info[-1]

            if interest < 1 :
                raise Exception("exception of our assumption that T converges to normal.")
            else:
                bucket.append(interest)
        buckets.append(bucket)
        print("                =>success")

    return buckets
buckets=compute_bucket(col=10,n=100,sample_size=10000,iter=5)
```

```python
def compute_plot(buckets,bins=50,xlim=100,ylim=0.2):
    #plotting
    iter=len(buckets)
    sample_size=len(buckets[0])
    fig, axes = plt.subplots(1,iter,sharex=True,sharey=True,figsize=(16,4))

    for idx in range(iter):

        axes[idx].hist(buckets[idx],density=True,bins=bins,color="magenta",histtype='bar',alpha=0.5,range=(0,100),edgecolor ="black")
        # Add labels and title
        axes[idx].set_xlabel('Proportions',rotation=0)
        if idx==0:
            axes[idx].set_ylabel('Probability',rotation=90)
        axes[idx].set_title("Plot {}".format(idx+1))
        axes[idx].set_xlim(0,xlim)
        axes[idx].set_ylim(0,ylim)
        axes[idx].vlines(ymin=0,ymax=1,x=1,colors='r',linestyles='--')
        axes[idx].grid(color='0.95',linestyle="--")
        axes[idx].hlines(xmin=0,xmax=100,y=0,colors='black',alpha=0.3,linestyles='--')
        print("generating plot.."+"({0}/{1})".format(idx+1,iter))

    plt.tight_layout()
    plt.show()
compute_plot(buckets,20,100,0.15)
```

```python
def compute_table(buckets):
    iter=len(buckets)
    sample_size=len(buckets[0])

    for idx in range(iter):
        print("{0:2}. min:{1:<10.3f} max:{2:<10.3f} mean:{3:<10.3f}".format(idx+1,min(buckets[idx]),max(buckets[idx]),sum(buckets[idx])/sample_size))
compute_table(buckets)
```
