## Convergence of the iterated mean transforms of a 2 × 2 matrix; its Algorithm

#### Ji Eun Lee and Min Woo Park

#### Abstract

For the polar decomposition T = U |T | ∈ L(H) (defined below), the mean
transform ̂ T of T is defined by
```
T = \frac{1}{2} (T + |T |U ).
```

We prove that the iterated mean transform of a 2 × 2 matrix converges
a normal matrix. Moreover, we provide the computation method of the
iterated mean transforms with an Algorithm in Python.

#### Example 2.4.
 If
$T=
\begin{pmatrix}
0&1\\
0&0
\end{pmatrix}$ on ${\mathbb C}^2$, then $T:=U|T|=\begin{pmatrix}
    0 & 1 \\
   0 & 0
    \end{pmatrix}\begin{pmatrix}
    0&0\\
    0& 1
    \end{pmatrix}$ is the polar decomposition of $T$.
Thus
$
    \widehat{T}=\frac{1}{2}(T+\widetilde{T}^{(D)})=\frac{1}{2}
    \begin{pmatrix}
    0&1\\
    0&0
    \end{pmatrix}
$
and so $\widehat{T}$ is normal.
Hence $\widehat{T}^{(n)}$ converges to normal as $n\rightarrow \infty$.

#### Example 2.5.

Let $T=
\begin{pmatrix}
0 & 1 \\
0 & 1
\end{pmatrix}$.
Then $T=U|T|=
\begin{pmatrix}
    0 & \frac{\sqrt{2}}{2} \\
   0 & \frac{\sqrt{2}}{2} 
    \end{pmatrix}\begin{pmatrix}
    0&0\\
   0& \sqrt{2}
    \end{pmatrix}$ is the polar decomposition of $T$ and $\widetilde{T}^D=
\begin{pmatrix}
    0&0\\
   0& \sqrt{2}
    \end{pmatrix}
\begin{pmatrix}
    0 & \frac{\sqrt{2}}{2} \\
   0 & \frac{\sqrt{2}}{2} 
    \end{pmatrix}=
\begin{pmatrix}
0 & 0\\
0 & 1\\ 
\end{pmatrix}$. Thus
\begin{align*}
    T_1=\widehat{T}=\frac{1}{2}\left[
    \begin{pmatrix}
0 & 1\\
0 & 1\\ 
\end{pmatrix}+
\begin{pmatrix}
0 & 0\\
0 & 1\\ 
\end{pmatrix}\right]=\frac{1}{2}\begin{pmatrix}
0 & 1\\
0 & 2\\ 
\end{pmatrix}=
\begin{pmatrix}
0 & \frac{1}{2}\\
0 & 1\\ 
\end{pmatrix}.
\end{align*}
And so
\begin{align*}
&|T_1|=(\widehat{T}^*\widehat{T})^\frac{1}{2}
    =\left[
    \begin{pmatrix}
0 & 0\\
\frac{1}{2} & 1\\ 
\end{pmatrix}
\begin{pmatrix}
0 & \frac{1}{2}\\
0 & 1\\ 
\end{pmatrix}\right]^\frac{1}{2}
=
\begin{pmatrix}
0 & 0\\
0 & \frac{\sqrt{5}}{2}\\ 
\end{pmatrix} and
&U_1=\frac{1}{\sqrt{5}}
\begin{pmatrix}
0 & 1\\
0 & 2
\end{pmatrix}.
\end{align*}
Therefore
\begin{align*}
    T_2=\widehat{T_1}&=
    \frac{1}{2}\left[
    \begin{pmatrix}
0 & \frac{1}{2}\\
0 & 1\\ 
\end{pmatrix}+
\begin{pmatrix}
0 & 0\\
0 & \frac{\sqrt{5}}{2}\\ 
\end{pmatrix}
\frac{1}{\sqrt{5}}
\begin{pmatrix}
0 & 1\\
0 & 2\\ 
\end{pmatrix}
\right]\\
&=\frac{1}{2}
\begin{pmatrix}
0 & \frac{1}{2}\\
0 & 2\\ 
\end{pmatrix}=
\begin{pmatrix}
0 & \left(\frac{1}{2}\right)^2\\
0 & 1\\ 
\end{pmatrix}.
\end{align*}

#### Example 2.9.

Let $T=
\begin{pmatrix}
i & 2 \\
0 & -i
\end{pmatrix}$.
Then
\begin{equation*}
    |T|^2=T^*T =
    \begin{pmatrix}
        -i & 0 \\
        2 & i
    \end{pmatrix}
    \begin{pmatrix}
        i & 2 \\
        0 & -i
    \end{pmatrix}
    =
    \begin{pmatrix}
        1 & -2i \\
        2i & 5
    \end{pmatrix}.
\end{equation*}
By Lemma 2.7, we have
\begin{equation*}
    |T|=\frac{1}{\sqrt{2}}
    \begin{pmatrix}
        1 & -i \\
        i & 3
    \end{pmatrix}~\text{and}~
    |T|^{-1}=\frac{1}{\sqrt{2}}
    \begin{pmatrix}
        3 & i \\
        -i & 1
    \end{pmatrix}.
\end{equation*}
Thus
\begin{align*}
    &U=T|T|^{-1}=
    \begin{pmatrix}
        i & 2 \\
        0 & -i
    \end{pmatrix}
    \frac{1}{\sqrt{2}}
    \begin{pmatrix}
        3 & i \\
        -i & 1
    \end{pmatrix}=
    \frac{1}{\sqrt{2}}
    \begin{pmatrix}
        i & 1 \\
        -1 & -i
    \end{pmatrix}\quad \text{and}\\
    &\widetilde{T}^D = |T|U =
    \frac{1}{\sqrt{2}}
    \begin{pmatrix}
        1 & -i \\
        i & 3
    \end{pmatrix}
    \frac{1}{\sqrt{2}}
    \begin{pmatrix}
        i & 1 \\
        -1 & -i
    \end{pmatrix}=
    \frac{1}{2}
    \begin{pmatrix}
        2i & 0 \\
        -4 & -2i
    \end{pmatrix}=
    \begin{pmatrix}
        i & 0 \\
        -2 & i
    \end{pmatrix}.
\end{align*}
Hence we get
\begin{equation*}
    \widehat{T}=\frac{1}{2}(T+\widetilde{T}^D)=\frac{1}{2}\left(
    \begin{pmatrix}
        i & 2 \\
        0 & -i
    \end{pmatrix}+
    \begin{pmatrix}
        i & 0 \\
        -2 & i
    \end{pmatrix}
    \right)
    =
    \begin{pmatrix}
        i & 1 \\
        -1 & -i
    \end{pmatrix}.
\end{equation*}
Since
\begin{align*}
    det(\widehat{T}-\lambda I_2 ) =
    \begin{vmatrix}
    i-\lambda & 1\\
    -1 & -1-\lambda
    \end{vmatrix}
    =(i-\lambda)(-i-\lambda)+1
    =\lambda^2+2
    =0,
\end{align*}
it follows that the eigenvalues of $\widehat{T}$ are $\pm \sqrt{2}i$.

To satisfy the given conditions, we can let:

$T=\begin{pmatrix} a & b \ -b & a \end{pmatrix}$

where $a,b\in\mathbb{R}$ satisfy $b > 0$ and $\det(T) = ad - bc = a^2 + b^2 = 1$.

Then, the characteristic equation of $T$ is:

$\det(T - \lambda I) = \begin{vmatrix} a-\lambda & b \ -b & a-\lambda \end{vmatrix} = (a-\lambda)^2 + b^2 = (\lambda - u - iv)(\lambda - u + iv)$

where $\lambda = u + iv$ and $\mu = u - iv$ are the eigenvalues of $T$.

Expanding the determinant, we get:

$(a-\lambda)^2 + b^2 = a^2 - 2a\lambda + \lambda^2 + b^2 = (\lambda - u)^2 + v^2 + b^2$

Equating the real and imaginary parts of this expression, we get:

$a^2 - 2au + u^2 + b^2 = (\lambda - u)^2 + b^2 = (\lambda - u)^2 + (\mu - u)^2$

$-2av + 2v\mu = 2v^2$

Simplifying, we get:

$2u^2 + 2v^2 - 2au = 0$

$v(\lambda - \mu) = 0$

Since $v > 0$, we must have $\lambda = u + iv$ and $\mu = u - iv$ as the two roots of the characteristic equation.

We can choose $a = \frac{1}{\sqrt{2}}$ and $b = \frac{1}{\sqrt{2}}$, then:

$T=\begin{pmatrix} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \ -\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \end{pmatrix}$

and we have:

$\lambda = u + iv = \frac{1}{\sqrt{2}} + i\frac{1}{\sqrt{2}}$

$\mu = u - iv = \frac{1}{\sqrt{2}} - i\frac{1}{\sqrt{2}}$

````python
a = np.sqrt(2)
T=np.matrix([1/a,a,0,complex(0,1/a)]).reshape(2,2)
# T = np.matrix([1,2,-2/3,1/3]).reshape(2, 2)
Transformer = MeanTrans()
Transformer.transform(T, 50)

Transformer.normal_calculator()
Transformer.normal_plot()


print(T@T.H-T.H@T)
print(sum(np.linalg.eig(T)[0]))

T_1=Transformer[1]
print(sum(np.linalg.eig(T_1)[0]))

print(np.linalg.eig(T)[0])
print(np.linalg.eig(T_1)[0])
````
